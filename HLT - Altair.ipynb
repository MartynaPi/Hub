```python
import altair as alt
alt.renderers.enable('altair_viewer')
```

```python
from vega_datasets import data  
iris = data.iris()              
iris.head()
```

```python
df = iris
chart = alt.Chart(df)
```

```python
df
```

```python
alt.Chart(df).mark_point()
```

```python
chart = alt.Chart(iris).mark_point().encode(
    x='sepalLength',
    y='sepalWidth',
    color='species')
chart.show()
```

```python
alt.Chart(iris).mark_bar().encode(
    x=alt.X('petalLength', bin=alt.Bin(maxbins=30)),
    y= 'count()',
    color= 'species'
)
```
