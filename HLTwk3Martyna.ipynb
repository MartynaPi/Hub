```python
import numpy as np

#1 Create a 1D array of numbers from 0 to 9
#or
b = np.arange (10)
print (b)

#2 Create a 3×3 NumPy array of allBoolean value Trues

np.random.randint(8, size=(3,3))

filedata = np.random.randint(8, size=(3,3))
print (filedata)
print (filedata < 8 )

#3 Extract all odd numbers from array of 1-10
c = np.arange (10)
print (c[c % 2 == 1])

#4 Replace all odd numbers in an array of 1-10 with the value -1

a = np.arange (10)
a [a % 2 == 1] = -1
print(a)

#5 Convert a 1D array to a 2D array with 2 rows
arr1 = np.array ([1,2,3,4,5,6])
print (arr1)
arr2 = arr1.reshape ((2,3))
print (arr2)

#6 Create two arrays a and b, stack these two arrays vertically use the  np.dot and np.sum to calculate totals
a1 = np.array([1,2,3,4])
a2 = np.array([5,6,7,8])

print (np.vstack([a1,a2]))
final = np.vstack([a1,a2])
print (np.sum(final, axis=0))


#7 Create the following pattern without hardcoding. Use only NumPy functions.
n = 8
arr1 = np.array([n-2, n-2, n-2, n-1, n-1, n-1, n, n, n, n-2, n-1, n, n-2, n-1, n])
print (arr1)

#8 In two arrays a ( 1,2,3,4,5) and b ( 4,5,6,7,8,9) – remove all repeating  items present in array b
a = np.array ([1,2,3,4,5])
b = np.array ([4,5,6,7,8,9])

print (np.setdiff1d(b,a))

#9 Get all items between 3 and 7 from a and b and sum them together
a = np.array ([1,2,3,4,5])
b = np.array ([4,5,6,7,8,9])
print (a[(a >= 3) & (a <= 7)])
print (b[(b >= 3) & (b <= 7)])
final = np.array([[3,4,5,0],[4,5,6,7]])
print (np.sum (final))
```
